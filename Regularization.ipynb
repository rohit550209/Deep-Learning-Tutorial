```python
import pandas as pd
import numpy as np

import warnings
warnings.filterwarnings("ignore")
```

```python


df = pd.read_csv("C:/Users/Dell/Desktop/Deep_Learning/Data/sonar_dataset.csv", header= None)
df.sample(10)
```

```python
df.isna().sum()
```

```python
df.columns
```

```python
df[60].value_counts()
```

```python
X = df.drop(60, axis= 1)
y = df[60]
```

```python
y = pd.get_dummies(df[60], drop_first= True)
y
```

```python
y.value_counts()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size= 0.2, random_state= 1) # default test_size = 0.25
```

```python
len(X_train), len(X_test)
```

```python
import tensorflow as tf
from tensorflow import keras
```

```python
model = keras.Sequential([
    keras.layers.Dense(60, input_dim = 60, activation= "relu"),
    keras.layers.Dense(30, activation= "relu"),
    keras.layers.Dense(15, activation= "relu"),
    keras.layers.Dense(1, activation= "sigmoid")
])

model.compile(
    optimizer = "adam",
    loss = "binary_crossentropy",
    metrics = ["accuracy"]
)

model.fit(X_train, y_train, epochs= 100, batch_size= 10)
```

```python
model.evaluate(X_test, y_test)
```

```python
y_pred = model.predict(X_test)
y_preds = y_pred.reshape(-1)
y_preds
```

```python
y_prediction = np.round(y_preds)
y_prediction
```

```python
from sklearn.metrics import classification_report

print(classification_report(y_test, y_prediction))
```

```python
model = keras.Sequential([
    keras.layers.Dense(60, input_dim = 60, activation= "relu"),
    keras.layers.Dropout(0.5),
    keras.layers.Dense(30, activation= "relu"),
    keras.layers.Dropout(0.5),
    keras.layers.Dense(15, activation= "relu"),
    keras.layers.Dropout(0.5),
    keras.layers.Dense(1, activation= "sigmoid")
])

model.compile(
    optimizer = "adam",
    loss = "binary_crossentropy",
    metrics = ["accuracy"]
)

model.fit(X_train, y_train, epochs= 100, batch_size= 10)
```

```python
model.evaluate(X_test, y_test) # Although the training accuracy is decreased, the test accuracy is increased
```

```python
y_prediction = np.round(model.predict(X_test))
print(classification_report(y_test, y_prediction))
```
