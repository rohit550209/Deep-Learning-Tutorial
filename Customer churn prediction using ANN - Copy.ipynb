```python
def print_col_unique_val(df2):
    for col in df2:
        print(f"{col}: {df2[col].unique()} Total : {len(df2[col].unique())}")  #printing unique values of each column
```

```python
print_col_unique_val(df2)
```

```python
df2.replace(["No phone service", "No internet service"] , "No", inplace= True)
```

```python
print_col_unique_val(df2)
```

```python
yes_no_columns = ['Partner','Dependents','PhoneService','MultipleLines','OnlineSecurity','OnlineBackup',
                  'DeviceProtection','TechSupport','StreamingTV','StreamingMovies','PaperlessBilling','Churn']
for col in yes_no_columns:
    df2[col].replace({"Yes": 1,"No":0}, inplace= True)
```

```python
print_col_unique_val(df2)
```

```python
df2["gender"].replace({"Male":1, "Female":0}, inplace= True)
```

```python
print_col_unique_val(df2)
```

```python
df2 = pd.get_dummies(data= df2, columns=["InternetService", "Contract", "PaymentMethod"])
```

```python
print_col_unique_val(df2)
```

```python
from sklearn.preprocessing import MinMaxScaler

cols_to_scale = ["tenure", "MonthlyCharges", "TotalCharges"]
scalar = MinMaxScaler()

df2[cols_to_scale] = scalar.fit_transform(df2[cols_to_scale])
```

```python
print_col_unique_val(df2)
```

```python
X = df2.drop("Churn", axis= 1)
y = df2["Churn"]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size= 0.2, random_state=1)
```

```python
X_train.shape
```

```python
y_train.shape
```

```python
import tensorflow
from tensorflow import keras

model = keras.Sequential([
    keras.layers.Dense(20, input_shape = (26,), activation= "relu"),
    keras.layers.Dense(18, activation= "relu"),
    keras.layers.Dense(1, activation= "sigmoid")
])

model.compile(
    optimizer = "adam",
    loss = "binary_crossentropy",
    metrics = ["accuracy"]
)

model.fit(X_train, y_train, epochs = 30)
```

```python
y_pred = model.predict(X_test)
y_pred[:10]
```

```python
y_pr = []
for i in range(len(y_pred)):
    if y_pred[i] >= 0.5:
        y_pr.append(1)
    else:
        y_pr.append(0)
```

```python
len(y_pred), len(y_pr)
```

```python
match = 0 
for i, j in zip(y_pr, y_test):
    print(i,j)
    if i == j:
        match +=1
```

```python
match 
```

```python
from sklearn.metrics import confusion_matrix, classification_report

print(classification_report(y_pr, y_test))
```

```python
cm = confusion_matrix(y_pr, y_test)
print(cm)
```

```python
import seaborn as sns

plt.figure(figsize=(5,4))
sns.heatmap(cm, annot= True, fmt = "d")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()
```
